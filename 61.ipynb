```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([1, 0])
x2 = np.array([0, 1])
t = np.array([0, 1])

alpha = 0.5
w1 = 0.5
w2 = 0.5
b=0.2
loss = []
for epoh in range(5) :
  loss.append(0)
  for NoData in range(2):
    net = x1[NoData]*w1+x2[NoData]*w2+b
    if net >= 0 :
      y = 1
    else:
      y = 0
    e = t[NoData]-y
    dw1 = e*alpha*x1[NoData]
    dw2 = e*alpha*x2[NoData]
    db = e*alpha

    w1 = w1+dw1
    w2 = w2+dw2
    b=b+db
    loss[epoh]=loss[epoh]+np.abs(e)
  plt.plot(loss)
```

```python
def TrainingPerceptron(x, T):
  JumlahData = x.shape[0]
  JumlahVektorInput = x.shape[1]
  JumlahVektorOutput = T.shape[1]
  w = np.random.rand(JumlahVektorInput, JumlahVektorOutput) * 0.5
  alpha = 0.5
  loss = []
  
  for epoh in range(1000):
    loss.append(0)
    
    for NData in range(JumlahData):
      xi = x[NData:NData+1, :]
      net = np.matmul(xi, w)
      y = 1/(1 + np.exp(-net))
      e = T[NData:NData+1,:] - y
      dw = alpha * np.matmul(xi.T, e)
      w = w + dw
      loss[epoh] += np.sum(np.abs(e))
    
  return w, loss

def Prediksi(xp, w):
  net = np.matmul(xp, w)
  yp = 1 / (1 + np.exp(-net))
  return yp

X = np.loadtxt("/content/XJamak.csv", delimiter=',')
T = np.loadtxt("/content/TJamak.csv", delimiter=',')

w, loss = TrainingPerceptron(X, T)

plt.plot(loss)
plt.show()

xp = np.array([[1, 0, 1, 1]])
yp = Prediksi(xp, w)
print("Hasil Prediksi:")
print(yp)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt

X = np.array([[0,0],
             [0,1],
             [1,0],
             [1,1]])

T = np.array([[1],
             [0],
             [0],
             [0]])

model = Sequential()

NNode = 1
NInput = 2
model.add(Dense(NNode, input_dim=NInput, activation='sigmoid', use_bias=True))

model.compile(loss='mse', optimizer='SGD', metrics=['accuracy'])

His = model.fit(X, T, epochs=100)

plt.plot(His.history['loss'])
plt.show()

yp = model.predict(X)
print("Hasil Prediksi:")
print(yp)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt

X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

T = np.array([[1],
              [0],
              [0],
              [0]])

model = Sequential()
NOutput = 1
NInput = 2
model.add(Dense(1000, input_dim=NInput, activation='relu', use_bias=True))
model.add(Dense(1000, activation='relu', use_bias=True))
model.add(Dense(1000, activation='relu', use_bias=True))
model.add(Dense(NOutput, activation='sigmoid', use_bias=True))

model.compile(loss='mse', optimizer='SGD', metrics=['accuracy'])

His = model.fit(X, T, epochs=100)

plt.plot(His.history['loss'])
plt.show()

yp = model.predict(X)
print("Hasil Prediksi:")
print(yp)
```
